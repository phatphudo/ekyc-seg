```python
_, y_max = max(enumerate(points), key=lambda x: (x[1][1], x[1][0])) # y_max
_, y_min = min(enumerate(points), key=lambda x: (x[1][1], x[1][0])) # y_min
_, x_max = max(enumerate(points), key=lambda x: (x[1][0], x[1][0])) # x_max
_, x_min = min(enumerate(points), key=lambda x: (x[1][0], x[1][0])) # x_min
```

```python
import math

def dist(point1, point2):
    x1, y1 = point1
    x2, y2 = point2
    distance = math.sqrt((x2 - x1)**2 + (y2 - y1)**2)
    return distance
```

```python
min_tl, min_tr, min_bl, min_br = math.inf, math.inf, math.inf, math.inf
```

```python
for point in points:
    if abs(point[0] - x_min[0]) + abs(point[1] - y_min[1]) < min_tl:
        tl = point
        min_tl = abs(point[0] - x_min[0]) + abs(point[1] - y_min[1])
    if abs(point[0] - x_max[0]) + abs(point[1] - y_min[1]) < min_tr:
        tr = point
        min_tr = abs(point[0] - x_max[0]) + abs(point[1] - y_min[1])
    if abs(point[0] - x_min[0]) + abs(point[1] - y_max[1]) < min_bl:
        bl = point
        min_bl = abs(point[0] - x_min[0]) + abs(point[1] - y_max[1])
    if abs(point[0] - x_max[0]) + abs(point[1] - y_max[1]) < min_br:
        br = point
        min_br = abs(point[0] - x_max[0]) + abs(point[1] - y_max[1])
```

```python
import numpy as np

widthA = np.sqrt(((br[0] - bl[0]) ** 2) + ((br[1] - bl[1]) ** 2))
widthB = np.sqrt(((tr[0] - tl[0]) ** 2) + ((tr[1] - tl[1]) ** 2))
# ...and now for the height of our new image
heightA = np.sqrt(((tr[0] - br[0]) ** 2) + ((tr[1] - br[1]) ** 2))
heightB = np.sqrt(((tl[0] - bl[0]) ** 2) + ((tl[1] - bl[1]) ** 2))
# take the maximum of the width and height values to reach
# our final dimensions
maxWidth = max(int(widthA), int(widthB))
maxHeight = max(int(heightA), int(heightB))
# construct our destination points which will be used to
# map the screen to a top-down, "birds eye" view
dst = np.array([
	[0, 0],
	[maxWidth, 0],
	[maxWidth, maxHeight],
	[0, maxHeight]], dtype = "float32")
# calculate the perspective transform matrix and warp
# the perspective to grab the screen
M = cv2.getPerspectiveTransform(src, dst)
warp = cv2.warpPerspective(img, M, (maxWidth, maxHeight))
```

```python
cv2.imwrite('data/warp_2.jpg', cv2.cvtColor(warp, cv2.COLOR_RGB2BGR))
```
